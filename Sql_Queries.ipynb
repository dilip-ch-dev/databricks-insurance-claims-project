```sql
-- Create our project catalog
CREATE CATALOG IF NOT EXISTS smart_claims_dev
COMMENT 'Car insurance claims automation project end - end';

-- Verify it was created
SHOW CATALOGS;
```

```sql
-- Use our new catalog
USE CATALOG smart_claims_dev;

-- Create Landing schema for raw files/volumes
CREATE SCHEMA IF NOT EXISTS landing
COMMENT 'Landing zone for raw files before Delta conversion';

-- Create Bronze schema for raw Delta tables
CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'Bronze layer - raw ingested data in Delta format';

-- Create Silver schema for cleaned data
CREATE SCHEMA IF NOT EXISTS silver
COMMENT 'Silver layer - cleaned, validated, standardized data';

-- Create Gold schema for analytics
CREATE SCHEMA IF NOT EXISTS gold
COMMENT 'Gold layer - aggregated analytics and business metrics';

-- Verify all schemas created
SHOW SCHEMAS IN smart_claims_dev;
```
